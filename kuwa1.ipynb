pandas_datareaderを使って日経平均の推移をグラフ化
--------------------------------------------

```python
import datetime
import pandas_datareader.data as web

start = datetime.datetime(1990,5,16)
end = datetime.datetime(2017,11,11)
nikkei225 = web.DataReader("NIKKEI225", "fred", start, end)

%matplotlib inline
nikkei225.plot()
```
